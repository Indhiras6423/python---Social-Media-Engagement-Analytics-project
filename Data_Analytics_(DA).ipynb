```python
# 1.Data Import & Setup
```

```python
import pandas as pd
import numpy as np
media_df = pd.read_csv("https://raw.githubusercontent.com/GeethaGunasekaran1/Dataset_rep/refs/heads/main/social_media_engagement_5000.csv")
media_df.head()
```

```python
media_df.dtypes
```

```python
media_df["posted_at"]=pd.to_datetime(media_df["posted_at"])
media_df.dtypes
```

```python
# 2. Data Cleaning
```

```python
 # Cleaning Missing Data
media_df.isna().sum()
```

```python
# Numeric Imputation (age)

media_df["age"]=media_df["age"].fillna(media_df["age"].mean())
```

```python
# category imputation

media_df["gender"]=media_df["gender"].fillna(media_df["gender"].mode()[0])
media_df["sentiment"]=media_df["sentiment"].fillna(media_df["sentiment"].mode()[0])
```

```python
# engagement imputation

media_df["likes"].fillna(0,inplace=True)
media_df["shares"].fillna(0,inplace=True)
media_df["comments"].fillna(0,inplace=True)
```

```python
media_df.isna().sum()
```

```python
# Identify duplicates
media_df.duplicated().sum()

  # There is no duplicate rows in this data
```

```python
# Data Formatting
media_df.dtypes
```

```python
cols=["gender","country","post_type","post_category","device_type","sentiment"]

media_df[cols]=media_df[cols].astype("category")

media_df["hashtags"]=media_df["hashtags"].astype("string")
media_df.dtypes
```

```python
media_df["gender"] = media_df["gender"].str.lower()
media_df["country"] =  media_df["country"].str.strip()
```

```python
media_df["post_type"].unique()
media_df["device_type"].unique()
media_df["post_category"].unique()

```
